```python
df1.TotalCharges= pd.to_numeric(df1.TotalCharges)
```

```python
for column in df1:
    if df1[column].dtypes == 'object':
        print(f'{column} : {df1[column].unique()}')
```

```python
df1.replace('Yes', 1, inplace = True)
```

```python
df1.replace('No', 0, inplace = True)
df1.replace('No internet service', 0, inplace = True)
df1.replace('No phone service', 0, inplace = True)
```

```python
df1.head()
```

```python
df1.shape
```

```python
df1['gender'].replace({'Female': 1, 'Male': 0}, inplace = True)
```

```python
df1 = pd.get_dummies(data = df1, columns = ['InternetService', 'Contract', 'PaymentMethod'])
```

```python
df1
```

```python
df1.dtypes
```

```python
cols_to_scale = ['tenure', 'MonthlyCharges', 'TotalCharges']
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
df1[cols_to_scale] = scaler.fit_transform(df1[cols_to_scale])
                                             
```

```python
df1.head()
```

```python
df1.columns
```

```python
X = df1.drop('Churn', axis= 1 )
```

```python
y = df1['Churn']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
import tensorflow as tf
from tensorflow import keras

model = keras.Sequential([
    keras.layers.Dense(20, input_shape = (26,), activation = 'relu'),
    keras.layers.Dense(1, activation = 'sigmoid')
])

model.compile(optimizer = 'adam',
             loss = 'binary_crossentropy',
             metrics = ['accuracy'])

model.fit(X_train, y_train, epochs = 100)
```

```python
model.evaluate(X_test, y_test)
```

```python
yp = model.predict(X_test)
yp[:5]
```

```python
y_test[:5]
```

```python
for i in range(len(yp)):
    if yp[i]>0.5:
        yp[i] = 1
    else:
        yp[i] = 0
        
```

```python
yp[:10]
```

```python
y_test[:10]
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
print(classification_report(y_test, yp))
```

```python
cm = tf.math.confusion_matrix(labels =y_test, predictions = yp)
sns.heatmap(cm, annot = True, fmt = 'd')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```
